```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Definicja sieci neuronowej
class PINN(nn.Module):
    def __init__(self):
        super(PINN, self).__init__()
        self.fc1 = nn.Linear(1, 20)
        self.fc2 = nn.Linear(20, 20)
        self.fc3 = nn.Linear(20, 1)
        self.activation = torch.tanh

    def forward(self, x):
        x = self.activation(self.fc1(x))
        x = self.activation(self.fc2(x))
        x = self.fc3(x)
        return x

# Ustawienia problemu
omega = 1.0  # Można zmienić na dowolną wartość
x_domain = np.linspace(-2*np.pi, 2*np.pi, 100).reshape(-1, 1)
x_train = torch.tensor(x_domain, dtype=torch.float32, requires_grad=True)

# Warunek początkowy
u_0 = torch.tensor([0.0], requires_grad=True)

# Funkcja kosztu
def loss_function(model, x_train, omega):
    u_pred = model(x_train)
    u_pred_x = torch.autograd.grad(u_pred, x_train, grad_outputs=torch.ones_like(u_pred), create_graph=True)[0]
    residual = u_pred_x - torch.cos(omega * x_train)
    residual_loss = torch.mean(residual ** 2)
    
    ic_pred = model(torch.tensor([[0.0]]))
    ic_loss = torch.mean((ic_pred - u_0) ** 2)
    
    return residual_loss + ic_loss

# Inicjalizacja modelu i optymalizatora
model = PINN()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Trening modelu
epochs = 50000
for epoch in range(epochs):
    optimizer.zero_grad()
    loss = loss_function(model, x_train, omega)
    loss.backward()
    optimizer.step()
    
    if epoch % 5000 == 0:
        print(f'Epoch {epoch}: Loss = {loss.item()}')

# Predykcje modelu
x_test = torch.tensor(x_domain, dtype=torch.float32)
u_pred = model(x_test).detach().numpy()

# Rozwiązanie analityczne
u_analytical = (1/omega) * np.sin(omega * x_domain)

# Wykres wyników
plt.figure(figsize=(10, 6))
plt.plot(x_domain, u_analytical, label='Analytical Solution', linestyle='dashed')
plt.plot(x_domain, u_pred, label='PINN Prediction')
plt.xlabel('x')
plt.ylabel('u(x)')
plt.legend()
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Definicja sieci neuronowej
class PINN(nn.Module):
    def __init__(self, layers):
        super(PINN, self).__init__()
        self.layers = nn.ModuleList()
        for i in range(len(layers) - 1):
            self.layers.append(nn.Linear(layers[i], layers[i+1]))
        self.activation = torch.tanh

    def forward(self, x):
        for i in range(len(self.layers) - 1):
            x = self.activation(self.layers[i](x))
        x = self.layers[-1](x)
        return x

# Funkcja kosztu
def loss_function(model, x_train, omega):
    u_pred = model(x_train)
    u_pred_x = torch.autograd.grad(u_pred, x_train, grad_outputs=torch.ones_like(u_pred), create_graph=True)[0]
    residual = u_pred_x - torch.cos(omega * x_train)
    residual_loss = torch.mean(residual ** 2)
    
    ic_pred = model(torch.tensor([[0.0]]))
    ic_loss = torch.mean((ic_pred - torch.tensor([0.0])) ** 2)
    
    return residual_loss + ic_loss

# Trening modelu
def train_model(model, x_train, omega, epochs=50000, lr=0.001):
    optimizer = optim.Adam(model.parameters(), lr=lr)
    loss_history = []

    for epoch in range(epochs):
        optimizer.zero_grad()
        loss = loss_function(model, x_train, omega)
        loss.backward()
        optimizer.step()
        loss_history.append(loss.item())
    
    return model, loss_history

# Wizualizacja wyników
def plot_results(x_domain, u_analytical, u_pred, loss_history, title):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(x_domain, u_analytical, label='Analytical Solution', linestyle='dashed')
    plt.plot(x_domain, u_pred, label='PINN Prediction')
    plt.xlabel('x')
    plt.ylabel('u(x)')
    plt.legend()
    plt.title(title)

    plt.subplot(1, 2, 2)
    plt.plot(range(len(loss_history)), loss_history)
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training Loss')
    plt.show()
```

```python
# Ustawienia problemu
omega = 1.0
x_train = np.linspace(-2*np.pi, 2*np.pi, 200).reshape(-1, 1)
x_test = np.linspace(-2*np.pi, 2*np.pi, 1000).reshape(-1, 1)

x_train_torch = torch.tensor(x_train, dtype=torch.float32, requires_grad=True)
x_test_torch = torch.tensor(x_test, dtype=torch.float32, requires_grad=True)

u_analytical = (1/omega) * np.sin(omega * x_test)
```

```python
# Definicja modelu
model = PINN(layers=[1, 16, 16, 1])

# Trening modelu
model, loss_history = train_model(model, x_train_torch, omega, epochs=50000, lr=0.001)

# Predykcje modelu
u_pred = model(x_test_torch).detach().numpy()

# Wizualizacja wyników
plot_results(x_test, u_analytical, u_pred, loss_history, title='Case (a): ω = 1, 2 hidden layers, 16 neurons each')
```

```python
class PINNAnsatz(nn.Module):
    def __init__(self, layers, omega):
        super(PINNAnsatz, self).__init__()
        self.omega = omega
        self.layers = nn.ModuleList()
        for i in range(len(layers) - 1):
            self.layers.append(nn.Linear(layers[i], layers[i+1]))
        self.activation = torch.tanh

    def forward(self, x):
        u = x
        for i in range(len(self.layers) - 1):
            u = self.activation(self.layers[i](u))
        u = self.layers[-1](u)
        return torch.tanh(self.omega * x) * u

# Trening modelu z Ansatz
def loss_function_ansatz(model, x_train, omega):
    u_pred = model(x_train)
    u_pred_x = torch.autograd.grad(u_pred, x_train, grad_outputs=torch.ones_like(u_pred), create_graph=True)[0]
    residual = u_pred_x - torch.cos(omega * x_train)
    residual_loss = torch.mean(residual ** 2)
    
    return residual_loss

def train_model_ansatz(model, x_train, omega, epochs=50000, lr=0.001):
    optimizer = optim.Adam(model.parameters(), lr=lr)
    loss_history = []

    for epoch in range(epochs):
        optimizer.zero_grad()
        loss = loss_function_ansatz(model, x_train, omega)
        loss.backward()
        optimizer.step()
        loss_history.append(loss.item())
    
    return model, loss_history

# Definicja modelu
model_ansatz = PINNAnsatz(layers=[1, 16, 16, 1], omega=omega)

# Trening modelu
model_ansatz, loss_history_ansatz = train_model_ansatz(model_ansatz, x_train_torch, omega, epochs=50000, lr=0.001)

# Predykcje modelu
u_pred_ansatz = model_ansatz(x_test_torch).detach().numpy()

# Wizualizacja wyników
plot_results(x_test, u_analytical, u_pred_ansatz, loss_history_ansatz, title='Case (c): Ansatz Model')
```

```python
class PINNFourier(nn.Module):
    def __init__(self, layers, L):
        super(PINNFourier, self).__init__()
        self.L = L
        self.fc1 = nn.Linear(2*L, layers[1])
        self.layers = nn.ModuleList([nn.Linear(layers[i], layers[i+1]) for i in range(1, len(layers)-1)])
        self.activation = torch.tanh

    def forward(self, x):
        gamma_x = torch.cat([torch.sin(k * x) for k in range(1, self.L+1)] + [torch.cos(k * x) for k in range(1, self.L+1)], dim=-1)
        u = self.activation(self.fc1(gamma_x))
        for layer in self.layers:
            u = self.activation(layer(u))
        return u


L = 8
model_fourier = PINNFourier(layers=[1, 16, 16, 1], L=L)
model_fourier, loss_history_fourier = train_model(model_fourier, x_train_torch, omega, epochs=50000, lr=0.001)
u_pred_fourier = model_fourier(x_test_torch).detach().numpy()
plot_results(x_test, u_analytical, u_pred_fourier, loss_history_fourier, title='Case (d): Fourier Features')
```
